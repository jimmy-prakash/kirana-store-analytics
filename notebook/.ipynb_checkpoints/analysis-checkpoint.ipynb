### 3A - Setup and Overview

In this section, we load the cleaned dataset and perform a quick structural overview.
The goal is to understand the number of records, data types, and value distributions before visual exploration.


```python
# Load the cleaned dataset
cleaned_path = "../data/sales_data_cleaned.csv"
df = pd.read_csv(cleaned_path, parse_dates=['Date'])

# Verify
df.info()
df.head()
```

```python
# Summary statistics for numeric columns
df.describe()

# Unique category and item counts
print("\nUnique Items:", df['Item'].nunique())
print("Unique Categories:", df['Category'].nunique())
```

### 3B - Visual Explorations

This section visualizes key sales insights from the Kirana Store data.
We focus on identifying high-performing products, revenue distribution, and short-term sales patterns.


```python
# Top 5 items by total revenue
top_items = (
    df.groupby('Item')['Total Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(8,5))
top_items.plot(kind='bar', color='skyblue')
plt.title('Top 5 Items by Total Revenue')
plt.xlabel('Item')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save before showing
plt.savefig("../outputs/top_items.png", bbox_inches='tight')
plt.show()
```

```python
# Revenue by category
category_revenue = (
    df.groupby('Category')['Total Revenue']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(6,6))
category_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Revenue Distribution by Category')
plt.ylabel('')

# Save before showing
plt.savefig("../outputs/top_items.png", bbox_inches='tight')
plt.show()
```

```python
# Aggregate daily revenue
daily_trend = (
    df.groupby('Date')['Total Revenue']
    .sum()
    .sort_index()
)

plt.figure(figsize=(10,5))
plt.plot(daily_trend.index, daily_trend.values, marker='o', linestyle='-', linewidth=2)
plt.title('Daily Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True, linestyle='--', alpha=0.6)

# Save before showing
plt.savefig("../outputs/top_items.png", bbox_inches='tight')
plt.show()
```

```python
# Weekly revenue trend
df['Week'] = df['Date'].dt.isocalendar().week
weekly_trend = df.groupby('Week')['Total Revenue'].sum()

plt.figure(figsize=(10,5))
plt.plot(weekly_trend.index, weekly_trend.values, marker='o', color='orange')
plt.title('Weekly Revenue Trend')
plt.xlabel('Week Number')
plt.ylabel('Total Revenue')
plt.grid(True, linestyle='--', alpha=0.6)

# Save before showing
plt.savefig("../outputs/top_items.png", bbox_inches='tight')
plt.show()
```
